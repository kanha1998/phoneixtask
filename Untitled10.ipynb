```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('https://go.phoenixrobotix.com/ml-test-2-q1-data')
```

```python
data.shape
```

```python
data.columns
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
X=data['Time']
Y=data['Temp. (Degree Celcius)']
Y1=[]
i=0
X1=[];
while(i<3487):
    Y1.append(Y[i])
    X1.append(i)
    i=i+1;
```

```python

Y1=np.array(Y1)
daydata=np.resize(Y1,(146,24));


```

```python
MAX=[];
MIN=[];
AVG=[];
for i in range(0,146):
    MAX.append(max(daydata[i]));
    MIN.append(min(daydata[i]));
    AVG.append(np.average(daydata[i]));
```

i used logistic regression of average,min and max of 24 hrs data

```python
xtrain=[];
xtrain=[[MAX[i], MIN[i] ,AVG[i]] for i in range(0,31)]
xtrain=np.array(xtrain).reshape(-1,3)
```

```python
i=0
y12am=[];
while(i<3487):
    y12am.append(Y1[i]);
    i=i+24;
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg=reg.fit(xtrain,y12am[:31])
```

```python
reg.score(xtrain,y12am[:31])
```

```python
xtest=[[MAX[i], MIN[i] ,AVG[i]] for i in range(31,145)]
xtest=np.array(xtest).reshape(-1,3)
```

```python
reg.score(xtest,y12am[31:145])
```

```python

plt.plot(X1[31:145],y12am[31:145])
plt.plot(X1[31:145],reg.predict(xtest),color='red')
```
